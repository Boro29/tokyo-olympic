```python
from pyspark.sql.functions import col

EntriesGender = EntriesGender.withColumn("Female", col("Female").cast("int")) \
                             .withColumn("Male", col("Male").cast("int")) \
                             .withColumn("Total", col("Total").cast("int"))

display(EntriesGender)
```

```python
from pyspark.sql.functions import col, sum

# Convert the Gold column to integer
medals = medals.withColumn("Gold", col("Gold").cast("int"))

# Group by TeamCountry and sum the Gold medals
gold_medals_by_country = medals.groupBy("TeamCountry").agg(sum("Gold").alias("TotalGoldMedals"))

# Sort the countries by the number of Gold medals in descending order
top_countries_with_gold_medals = gold_medals_by_country.orderBy(col("TotalGoldMedals").desc())

display(top_countries_with_gold_medals)
```

```python
from pyspark.sql.functions import avg, sum

avg_entries_by_gender = EntriesGender.groupBy("Discipline") \
                                     .agg(avg("Female").alias("AvgFemaleEntries"),
                                          avg("Male").alias("AvgMaleEntries"),
                                          sum(col("Female") + col("Male")).alias("TotalEntries"))

display(avg_entries_by_gender)
```

```python
from pyspark.sql.functions import col, count

# Convert the Gold, Silver, and Bronze columns to integer
medals = medals.withColumn("Gold", col("Gold").cast("int")) \
               .withColumn("Silver", col("Silver").cast("int")) \
               .withColumn("Bronze", col("Bronze").cast("int"))

# Aggregate medals won by athletes
athlete_medals = athletes.join(medals, athletes.Country == medals.TeamCountry) \
                         .groupBy("PersonName", "Discipline") \
                         .agg(sum("Gold").alias("GoldMedals"),
                              sum("Silver").alias("SilverMedals"),
                              sum("Bronze").alias("BronzeMedals"))

# Aggregate medals won by coaches
coach_medals = coaches.join(medals, coaches.Country == medals.TeamCountry) \
                      .groupBy("Name", "Discipline") \
                      .agg(sum("Gold").alias("GoldMedals"),
                           sum("Silver").alias("SilverMedals"),
                           sum("Bronze").alias("BronzeMedals"))

# Union the two DataFrames
medals_won = athlete_medals.select(col("PersonName").alias("Name"), "Discipline", "GoldMedals", "SilverMedals", "BronzeMedals") \
                           .union(coach_medals.select("Name", "Discipline", "GoldMedals", "SilverMedals", "BronzeMedals"))

display(medals_won)
```

```python
import matplotlib.pyplot as plt
import squarify

# Aggregate total medals
total_medals = medals_won.groupBy().sum("GoldMedals", "SilverMedals", "BronzeMedals").collect()[0]
medal_counts = [total_medals["sum(GoldMedals)"], total_medals["sum(SilverMedals)"], total_medals["sum(BronzeMedals)"]]
medal_labels = ["Gold", "Silver", "Bronze"]

# Pie chart
plt.figure(figsize=(10, 6))
plt.pie(medal_counts, labels=medal_labels, autopct='%1.1f%%', colors=['#FFD700', '#C0C0C0', '#CD7F32'])
plt.title('Distribution of Medals')
plt.show()

# Tree map
plt.figure(figsize=(10, 6))
squarify.plot(sizes=medal_counts, label=medal_labels, color=['#FFD700', '#C0C0C0', '#CD7F32'], alpha=.8)
plt.title('Tree Map of Medals')
plt.axis('off')
plt.show()
```
